```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tv1 = pd.read_csv("./data/torchvision_bs1.csv").reset_index().rename(columns={'index': 'library'})
tv1['library'] = 'TorchVision'

kor1 = pd.read_csv("./data/kornia_bs1.csv").reset_index().rename(columns={'index': 'library'})
kor1['library'] = 'Kornia'

res = pd.concat([tv1, kor1]).set_index(['op_name', 'library'])
res.rename(columns=lambda x: x[-2:], inplace=True)
res.sort_index(inplace=True)
res = res.round(2)

print(res.to_latex())
```

```python
kor1 = pd.read_csv("./data/kornia_bs1.csv").reset_index().rename(columns={'index': 'batch_size'}).set_index('op_name')
kor1['batch_size'] = 1
kor2 = pd.read_csv("./data/kornia_bs2.csv").reset_index().rename(columns={'index': 'batch_size'}).set_index('op_name') / 2
kor2['batch_size'] = 2
kor4 = pd.read_csv("./data/kornia_bs4.csv").reset_index().rename(columns={'index': 'batch_size'}).set_index('op_name') / 4
kor4['batch_size'] = 4
kor8 = pd.read_csv("./data/kornia_bs8.csv").reset_index().rename(columns={'index': 'batch_size'}).set_index('op_name') / 8
kor8['batch_size'] = 8
kor16 = pd.read_csv("./data/kornia_bs16.csv").reset_index().rename(columns={'index': 'batch_size'}).set_index('op_name') / 16
kor16['batch_size'] = 16
kor32 = pd.read_csv("./data/kornia_bs32.csv").reset_index().rename(columns={'index': 'batch_size'}).set_index('op_name') / 32
kor32['batch_size'] = 32
kor64 = pd.read_csv("./data/kornia_bs64.csv").reset_index().rename(columns={'index': 'batch_size'}).set_index('op_name') / 64
kor64['batch_size'] = 64

merged = pd.concat([kor1, kor2, kor4, kor8, kor16, kor32, kor64]).reset_index().set_index(['op_name', 'batch_size']).sort_index()
```

```python
eff_index = 0
plt.figure(figsize=(8, 8))
for op in list(kor1.index):
    selected = merged.iloc[merged.index.get_level_values('op_name') == op].reset_index()[['batch_size', f'efficientnet-b{eff_index}']]
    v = selected[f'efficientnet-b{eff_index}'].values
    if op == 'ColorJitter':
        v /= 4
    plt.plot(["1", "2", "4", "8", "16", "32", "64"], v, label = op)

plt.ylim(ymin=0, ymax=4)
plt.xlim(xmin=0, xmax=6)
plt.xlabel('Batchsize', fontsize=12)
plt.ylabel('Time Cost (ms)', fontsize=12)
plt.legend(bbox_to_anchor=(1, 1))
plt.grid()
plt.savefig("./op_benchmarking.png")
```
